```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression, make_blobs
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

print("Libraries Imported")
```

```python
# 1. Generate Synthetic Data for Regression
# n_samples=100: We create 100 data points
# n_features=1: We have 1 input variable
# noise=15: Adds randomness so it's not a perfect line
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# 2. Convert to DataFrame for display purposes (Optional, helps visualization)
df_reg = pd.DataFrame(X_reg, columns=['Input (X)'])
df_reg['Target (y)'] = y_reg

# 3. Show the first 5 rows of the created dataset
print("--- Sample of Created Regression Data ---")
print(df_reg.head())
```
